# C) iii) Finding Optimum S

First, we collected data by varying the value of S from 0 to 100 and measuring the corresponding number of key comparisons and runtime for different data sizes.

We then plotted the smooth curve to understand the relationship between S and the number of key comparisons.

```python

try:
    data = np.loadtxt('varyingS.txt')
except Exception as e:
    print(f"Error loading data: {e}")
    exit()


x = data[:, 0]
y = data[:, 1]

x_bin_size = 1  # 1
y_bin_size = 1_000_000  # 1 million

x_scaled = x / x_bin_size
y_scaled = y / y_bin_size

sorted_indices = np.argsort(x_scaled)
x_scaled_sorted = x_scaled[sorted_indices]
y_scaled_sorted = y_scaled[sorted_indices]

X_linear_scaled = np.linspace(x_scaled_sorted.min(), x_scaled_sorted.max(), 500)

X_Y_Spline = make_interp_spline(x_scaled_sorted, y_scaled_sorted, k=3)  # k=3 for cubic spline

Y_linear_scaled = X_Y_Spline(X_linear_scaled)

# Plot the smooth curve
plt.figure(figsize=(10, 6))
plt.plot(X_linear_scaled * x_bin_size, Y_linear_scaled * y_bin_size, label="Smooth Curve", color='b')  # Scale back x and y values for plotting
#plt.scatter(x, y, color='r', s=10, label="Data Points")  # Original data points

def format_x_func(value, tick_number):
    return f'{int(value / x_bin_size)}'

def format_y_func(value, tick_number):
    return f'{int(value / y_bin_size)}M'

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_x_func))

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_func))

plt.xlim(x.min(), x.max())  # Keep the original x limits
plt.ylim(min(y) - y_bin_size, max(y) + y_bin_size)  # Add padding to y-axis limits

plt.title("Varying S vs Key Comparisons")
plt.xlabel("S")
plt.ylabel("Key Comparisons")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = np.loadtxt('size_S_comparisons_runtime.txt')

sizes = data[:, 0]         # First column: Sizes
S_values = data[:, 1]      # Second column: S (Threshold)
runtimes = data[:, 3]      # Fourth column: Runtime (in μs)

runtimes = np.round(runtimes / 10) * 10

unique_sizes = np.unique(sizes)
num_sizes = len(unique_sizes)

fig, axs = plt.subplots(num_sizes, 1, figsize=(12, 3 * num_sizes))
fig.suptitle('Runtime vs S for Different Sizes', fontsize=16)

for i, size in enumerate(unique_sizes):
    mask = sizes == size
    axs[i].plot(S_values[mask], runtimes[mask], markersize=5, linestyle='-', linewidth=1, color='blue')
    axs[i].set_title(f'Size {int(size)}')
    axs[i].set_xlabel('S (Threshold)')
    axs[i].set_ylabel('Runtime (μs)')
    axs[i].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

We plotted the smooth curve to understand the relationship between S and the number of key comparisons to show abrupt changes and potential areas of optimum S value.

```python
sizes = data[:, 0]    # First column: Sizes
S_values = data[:, 1]  # Second column: S values
runtimes = data[:, 3]  # Fourth column: Runtime (in μs)

runtimes = np.round(runtimes / 10) * 10

s_range = (30, 40)

fig, axes = plt.subplots(3, 2, figsize=(12, 12))
fig.suptitle("Runtime vs S Values (Range: 30-40)", fontsize=16)

unique_sizes = np.unique(sizes)

optimal_S_info = []

for i, size in enumerate(unique_sizes):
    mask = sizes == size
    current_S_values = S_values[mask]
    current_runtime = runtimes[mask]

    range_mask = (current_S_values >= s_range[0]) & (current_S_values <= s_range[1])
    filtered_S = current_S_values[range_mask]
    filtered_runtime = current_runtime[range_mask]

    if len(filtered_S) > 0:
        new_x = np.linspace(filtered_S.min(), filtered_S.max(), 100)

        poly_runtime = np.poly1d(np.polyfit(filtered_S, filtered_runtime, 2))

        new_y_runtime = poly_runtime(new_x)

        optimal_index = np.argmin(new_y_runtime)
        optimal_S = new_x[optimal_index]
        min_runtime = new_y_runtime[optimal_index]
        optimal_S_info.append((optimal_S, min_runtime))

        axes[i // 2, i % 2].plot(new_x, new_y_runtime, label=f"{int(size)} Runtime", color='yellow')
        axes[i // 2, i % 2].scatter(filtered_S, filtered_runtime, color='blue')

        axes[i // 2, i % 2].plot(filtered_S, filtered_runtime, color='blue')  # Connect dots with a dashed line

        axes[i // 2, i % 2].set_title(f"n = {int(size)}")
        axes[i // 2, i % 2].set_xlabel("S")
        axes[i // 2, i % 2].set_ylabel("Runtime (μs)")
        axes[i // 2, i % 2].legend()

axes[2, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 1.0])
plt.show()

for size, (s_value, runtime) in zip(unique_sizes, optimal_S_info):
    print(f"The optimal S value for n = {int(size)} is: {round(s_value)} with minimum runtime: {runtime:.2f} μs")

if optimal_S_info:
    common_optimal_S = np.mean([s_value for s_value, _ in optimal_S_info])
    print(f"The common optimal S value across all sizes is: {round(common_optimal_S)}")
```

For each size, we plotted runtime against S to identify any patterns or trends. The different sizes we used were n=1000. n=10000, n=100000, n=1000000, and n=10000000 and narrowed the range of s between 30-40

We found the optimum S value for each data size and averaged it out to find the overall optimum S as 34.
