```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator, FuncFormatter

# Load data
try:
    data = np.loadtxt('varyingNvsKeyComparisons.txt')
except Exception as e:
    print(f"Error loading data: {e}")
    exit()

#data = np.loadtxt('/varyingNvsKeyComparisons.txt')

# Split data into two columns
x = data[:, 0]
y = data[:, 1]

# Define the bin sizes
x_bin_size = 1_000_000  # 1 million
y_bin_size = 10_000_000  # 10 million

# Scale the x and y values
x_scaled = x / x_bin_size
y_scaled = y / y_bin_size

# Sort data by scaled x-values
sorted_indices = np.argsort(x_scaled)
x_scaled_sorted = x_scaled[sorted_indices]
y_scaled_sorted = y_scaled[sorted_indices]

# Create linearly spaced x-values within the range of scaled x-values
X_linear_scaled = np.linspace(x_scaled_sorted.min(), x_scaled_sorted.max(), 500)

# Create a spline interpolation of the data
X_Y_Spline = make_interp_spline(x_scaled_sorted, y_scaled_sorted, k=3)  # k=3 for cubic spline

# Evaluate the spline at the linearly spaced x-values
Y_linear_scaled = X_Y_Spline(X_linear_scaled)

# Plot the smooth curve
plt.figure(figsize=(10, 6))
plt.plot(X_linear_scaled * x_bin_size, Y_linear_scaled * y_bin_size, label="Smooth Curve", color='b')  # Scale back x and y values for plotting
#plt.scatter(x, y, color='r', s=10, label="Data Points")  # Original data points

# Custom formatter function to format the x-axis labels
def format_x_func(value, tick_number):
    return f'{int(value / x_bin_size)}M'

# Custom formatter function to format the y-axis labels
def format_y_func(value, tick_number):
    return f'{int(value / y_bin_size)}M'

# Set custom ticks on the x-axis
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_x_func))

# Set custom ticks on the y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_func))

# Set ticks to only full numbers for both axes
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))

# Optionally adjust the x-axis and y-axis limits
plt.xlim(x.min(), x.max())  # Keep the original x limits
plt.ylim(min(y) - y_bin_size, max(y) + y_bin_size)  # Add padding to y-axis limits

plt.title("Varying N vs Key Comparisons")
plt.xlabel("N")
plt.ylabel("Key Comparisons)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
try:
    data = np.loadtxt('1kn.txt')
    data1 = np.loadtxt('10kn.txt')
    data2 = np.loadtxt('100kn.txt')
    data3 = np.loadtxt('1miln.txt')
    data4 = np.loadtxt('10miln.txt')
except Exception as e:
    print(f"Error loading data: {e}")
    exit()


# Split data into two columns
x = data[:, 0]
f1kn = data[:, 1]
f10kn = data1[:, 1]
f100kn = data2[:, 1]
f1miln = data3[:, 1]
f10miln = data4[:, 1]

new_x=np.linspace(2,10,50)

poly1=np.poly1d(np.polyfit(x, f1kn, 2))
poly2=np.poly1d(np.polyfit(x, f10kn, 2))
poly3=np.poly1d(np.polyfit(x, f100kn, 2))
poly4=np.poly1d(np.polyfit(x, f1miln, 2))
poly5=np.poly1d(np.polyfit(x, f10miln, 2))

new_y1=poly1(new_x)
new_y2=poly2(new_x)
new_y3=poly3(new_x)
new_y4=poly4(new_x)
new_y5=poly5(new_x)

f,axes = plt.subplots(3,2,figsize=(10,10))
axes[0,0].plot(x,f1kn,linewidth=3)
axes[0,0].plot(new_x, new_y1)
axes[0,1].plot(x,f10kn,linewidth=3)
axes[0,1].plot(new_x, new_y2)
axes[1,0].plot(x,f100kn,linewidth=3)
axes[1,0].plot(new_x, new_y3)
axes[1,1].plot(x,f1miln,linewidth=3)
axes[1,1].plot(new_x, new_y4)
axes[2,0].plot(x,f10miln,linewidth=3)
axes[2,0].plot(new_x, new_y5)
plt.show()

#optS_1000=f1kn.index(min(f1kn))+1
#print("The optimal S value for n = 1000 is : ", end="")
indices = np.where(f1kn == min(f1kn))
#print(min(f1kn))
print("The optimal S value for n = 1000 is : " + str(indices[0]+1))

indices1 = np.where(f10kn == min(f10kn))
print("The optimal S value for n = 10,000 is : " + str(indices1[0]+1))

indices2 = np.where(f100kn == min(f100kn))
print("The optimal S value for n = 100,000 is : " + str(indices2[0]+1))

indices3 = np.where(f1miln == min(f1miln))
print("The optimal S value for n = 1,000,000 is : " + str(indices3[0]+1))

indices3 = np.where(f10miln == min(f10miln))
print("The optimal S value for n = 10,000.000 is : " + str(indices3[0]+1))

print("Average Optimal S Value is: ", ((indices[0]+1)+(indices1[0]+1)+(indices2[0]+1)+(indices3[0]+1)+(indices3[0]+1))//5)
```
